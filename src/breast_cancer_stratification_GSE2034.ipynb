### Outcome-Proximal Variables, Predictor Eligibility, and Leakage Control

To prevent semantic target leakage and ensure meaningful unsupervised stratification, predictor eligibility was determined based on **conceptual proximity to the outcome** rather than purely statistical association. At this stage of the analysis, the exact variable names in the dataset are not fully specified, so the focus is on **types of variables** that could introduce leakage.

Variables considered **outcome-proximal**, those that directly encode or reflect **relapse events** or **post-relapse information**, were excluded from the set of features used for stratification. Examples include:

- Binary indicators of distant or organ-specific metastasis  
- Time-to-event variables such as time to relapse or last follow-up  
- Any variables derived from post-relapse measurements

All remaining variables were retained for clustering and stratification, for instance:

- **Gene expression profiles (Affymetrix U133A microarray)** – baseline molecular features suitable for unsupervised analysis.  
- **Baseline clinical attributes** – e.g., receptor status or other pre-treatment measurements.  
- **Sample identifiers** – retained for data linkage but not used as predictive features.

By filtering variables in this way, stratification clusters are based solely on **baseline molecular and clinical characteristics**, avoiding circular definitions that could artificially inflate apparent differences in relapse outcomes. Outcome variables are reserved for **post-hoc evaluation** of cluster prognostic separation.

### Extracting Clinical Metadata from GEO (Gene Expression Omnibus)

The expression data we loaded previously contains the gene expression matrix, with **probes/genes as rows** and **samples (GSM IDs) as columns**. However, it does not include detailed information about the patients or experimental conditions.  

To analyze the dataset effectively, we also need the **metadata (clinical data)** for each sample, which typically includes:

- **Patient characteristics**: age, gender, clinical outcomes (e.g., relapse-free survival)  
- **Sample information**: tissue type, preparation method, batch, etc.  
- **Experimental details**: platform used, treatment, and other experimental annotations  

In GEO, this metadata is available in the same Series Matrix file, but lines starting with `!` are considered annotations and are skipped when loading the expression matrix directly with `pandas`.  

To access these annotations in a structured way, we use the **GEOparse** library. It allows us to download the GEO series and extract a **DataFrame of metadata**, one row per sample, with columns corresponding to all available annotation fields.  

This `df_metadata` will later be **merged with the `df_expression_data`** for downstream modeling and analysis.


```python
import GEOparse
import pandas as pd

# Download and parse the GEO Series Matrix
gse = GEOparse.get_GEO("GSE2034", destdir="./")

# Extract metadata for all samples
metadata = {gsm: gse.gsms[gsm].metadata for gsm in gse.gsms}

# Flatten metadata into a DataFrame
df_metadata = pd.DataFrame.from_dict(metadata, orient='index')

# Inspect the first 5 rows of clinical data
print(df_metadata.shape)
df_metadata.head()
```

## Target Leakage Prevention

To ensure valid prognostic stratification reflecting **baseline tumor biology only**, variables not plausibly available at the time of molecular profiling were excluded from clustering features prior to any preprocessing or feature engineering.

**Hard Leakage Variables (Reserved for Post-Hoc Analysis):** Outcome variables directly encoding relapse events were extracted from `characteristics_ch1` and **reserved exclusively** for survival analysis:

- `relapse:yes/no` event indicators
- `DFS_months` / time-to-relapse
- Distant metastasis status
- Post-treatment outcome annotations

These variables directly encode downstream survival outcomes and would invalidate unsupervised cluster discovery if used as clustering features.

**Soft Leakage Variables (Excluded from Clustering):** Clinically derived variables potentially contaminated by outcome knowledge were conservatively excluded:

- Any prognostic risk scores or indices from `characteristics_ch1`
- Post-profiling treatment response annotations  
- Derived clinical summaries dependent on follow-up data

Although some clinical descriptors (ER/PR status) may be partially available at baseline, their finalized annotations often incorporate longitudinal outcome data. Excluding them prevents reliance on downstream clinical summaries and ensures clusters reflect **pure pretreatment molecular heterogeneity**.

**Retained Features for Clustering:** Primary **gene expression matrix** (22,283 Affymetrix HG-U133A probes) + baseline sample descriptors only (`GSM_ID` for linkage). Survival endpoints reserved for Cox proportional hazards and Kaplan-Meier evaluation **after** molecular stratification.


```python
# Copy the original data to clean copies to preserve the original datasets
df_expression_data_clean = df_expression_data.copy()
df_metadata_clean = df_metadata.copy()

# Hard leakage: Features that should never be available at admission
hard_leakage_cols = [
    "Relapse Status",  # Example: this would be a post-relapse feature
    "Time to Relapse",  # Post-event information
    "Last Follow-Up",   # Information collected after treatment
    "Post-Treatment Response",  # Post-treatment info
    "Metastasis",  # Later-stage metastasis details
    "Disease Progression"  # Future progression after initial treatment
]

# Soft leakage: Typically determined after admission (conservative choice)
soft_leakage_cols = [
    "Tumor Stage",  # Could be determined post-admission, early stages onsidered
    "Survival Time",  # Time to event, should only be used for post-hoc analysis
    "Chemotherapy Type",  # Known after admission
    "Hormone Therapy Status",  # Typically determined after admission
    "Tumor Size",  # Might not be fully known at admission
]

# Remove hard leakage columns from df_expression_data (gene expression data)
df_expression_data_clean = df_expression_data_clean.drop(
    columns=[
        col for col in hard_leakage_cols if col in df_expression_data_clean.columns],
    errors="ignore"
)
# Remove **hard leakage** columns from df_metadata (metadata)
df_metadata_clean = df_metadata_clean.drop(
    columns=[
        col for col in hard_leakage_cols if col in df_metadata_clean.columns],
    errors="ignore"
)

# Remove soft leakage columns from df_expression_data (gene expression data)
df_expression_data_clean = df_expression_data_clean.drop(
    columns=[
        col for col in soft_leakage_cols if col in df_expression_data_clean.columns],
    errors="ignore"
)

# Remove **soft leakage** columns from df_metadata (metadata)
df_metadata_clean = df_metadata_clean.drop(
    columns=[
        col for col in soft_leakage_cols if col in df_metadata_clean.columns],
    errors="ignore"
)

# Check the cleaned datasets
print(f"Shape of df_expression_data after removing leakage columns: {
    df_expression_data_clean.shape}")
print(f"Shape of df_metadata after removing leakage columns: {
    df_metadata_clean.shape}")
df_expression_data_clean.head(), df_metadata_clean.head()
```
